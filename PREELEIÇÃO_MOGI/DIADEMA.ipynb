```python
# @title
from ntscraper import Nitter
import multiprocessing
import pandas as pd
```

```python
# @title
scraper = Nitter(0)
```

```python
tweets = scraper.get_tweets("mogidascruzes", mode = 'hashtag', number=1000)

final_tweets = []
for x in tweets['tweets']:
    data = [x['link'], x['text'],x['date'],x['stats']['likes'],x['stats']['comments']]
    final_tweets.append(data)

dat = pd.DataFrame(final_tweets, columns =['twitter_link','text','date','likes','comments'])
dat
```

```python
# Lista de palavras-chave
filtros = [
    'reeleição', 'candidato', 'partido', 'política',
    'candidato a prefeito', 'eleição', 'eleição 2024', 'pré candidatura', 'precandidato', 'prefeito apoia', 'aliança', 'pré-candidatos a prefeito', 'reeleição Caio Cunha'
]

# Função para verificar se o texto contém alguma palavra-chave e retornar a palavra-chave encontrada
def encontrar_filtro(texto):
    for filtro in filtros:
        if filtro in texto.lower():
            return filtro
    return None

# Aplicar a função para criar a nova coluna 'filtro' no DataFrame
dat['filtro'] = dat['text'].apply(encontrar_filtro)

# Exibir as linhas onde a coluna 'filtro' não é nula (ou seja, onde uma palavra-chave foi encontrada)
tweets_com_filtros = dat[dat['filtro'].notnull()]

# Exibir apenas as colunas relevantes
tweets_com_filtros = tweets_com_filtros[['twitter_link', 'text', 'date', 'likes', 'comments', 'filtro']]
tweets_com_filtros
```

```python
# Contar a frequência das palavras-chave na coluna 'filtro'
frequencia_filtros = tweets_com_filtros['filtro'].value_counts().head(5)

print(frequencia_filtros)
```

Analise da prefeitura de Mogi

```python
prefeitura = scraper.get_tweets("prefeituramogi", mode='user', number=1000)

# Estruturando os tweets obtidos em um DataFrame
final_prefeitura = []
for x in prefeitura['tweets']:
    data = [x['link'], x['text'], x['date'], x['stats']['likes'], x['stats']['comments']]
    final_prefeitura.append(data)

df_prefeitura = pd.DataFrame(final_prefeitura, columns=['twitter_link', 'text', 'date', 'likes', 'comments'])
df_prefeitura
```

```python
# Função para verificar se o texto contém alguma palavra-chave e retornar a palavra-chave encontrada
def encontrar_filtro_prefeitura(texto):
    for filtro in filtros:
        if filtro in texto.lower():
            return filtro
    return None

# Aplicar a função para criar a nova coluna 'filtro' no DataFrame df_prefeitura
df_prefeitura['filtro'] = df_prefeitura['text'].apply(encontrar_filtro_prefeitura)

# Exibir as linhas onde a coluna 'filtro' não é nula (ou seja, onde uma palavra-chave foi encontrada)
prefeitura_filtros = df_prefeitura[df_prefeitura['filtro'].notnull()]

# Exibir apenas as colunas relevantes
prefeitura_filtros = prefeitura_filtros[['twitter_link', 'text', 'date', 'likes', 'comments', 'filtro']]
prefeitura_filtros
```

```python
# Contar a frequência das palavras-chave na coluna 'filtro' para a prefeitura
frequencia_filtros_prefeitura = prefeitura_filtros['filtro'].value_counts()

print(frequencia_filtros_prefeitura)
```

```python
# Identificar as 10 linhas com mais likes no DataFrame df_prefeitura
top_likes = df_prefeitura.nlargest(10, 'likes')[['twitter_link', 'text', 'date', 'likes', 'comments']]

# Identificar as 10 linhas com mais comments no DataFrame df_prefeitura
top_comments = df_prefeitura.nlargest(10, 'comments')[['twitter_link', 'text', 'date', 'likes', 'comments']]

# Exibir as tabelas das 10 linhas com mais likes e comments
df_top_likes = pd.DataFrame(top_likes)
df_top_comments = pd.DataFrame(top_comments)
```
